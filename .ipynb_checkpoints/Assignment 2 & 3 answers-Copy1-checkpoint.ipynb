```python
#1. Show a breakdown of distance from home by job role and attrition
df.groupby(['JobRole','Attrition'])['DistanceFromHome'].mean()
```

```python
df.info() # called info() to check columns
```

```python
#2. Compare average monthly income by education and attrition
df.groupby(['Education','Attrition'])['MonthlyIncome'].mean()

```

```python
#Regardless of the years of Education employees with low monthly income tend to resign more often than those with higher income
```

```python
sns.set(style = "dark")
sns.displot(x='DistanceFromHome',data=df, hue='Attrition',col='JobRole')
```

```python
#Research Directors have the lowest attrition rate based on distance from home,
```

showing Monthy income in relation to Education and Attrition using seaborn

```python
sns.set(style = "white")
sns.displot(x='MonthlyIncome',data=df, hue='Attrition',col='Education')
```

```python
# checking for normal distribution
sns.displot(x='MonthlyIncome',data=df, hue='Attrition',col='Education', kind='kde')
```

#Generally, the rate of attrition is lower across all education groups, staff with many years of education tends to have the lowest attrition rate, possibly due to higher salaries.
